```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset('titanic')
```

```python
dataset.head(5)
```

```python
dataset.info()
```

```python
sns.histplot(x=dataset['fare'] , bins=30, kde= True)
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.jointplot(x=dataset['age'] ,y =dataset['fare'], kind='scatter')
sns.jointplot(x=dataset['age'] ,y = dataset['fare'], kind='hex')
```

```python
sns.rugplot(dataset['fare'])
```

```python
sns.barplot(x='sex', y='age', data= dataset, estimator= np.std)
```

```python
sns.countplot(x='sex', data= dataset)
```

```python
sns.boxplot(x='sex', y='age', data=dataset)
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=dataset)
```

```python
sns.violinplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
sns.stripplot(x='sex', y='age', data = dataset, jitter=True, hue='survived')
```

```python
sns.swarmplot(x='sex', y='age', data= dataset, hue='survived')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['sex']=le.fit_transform(dataset['sex'])
dataset['embarked']=le.fit_transform(dataset['embarked'])
dataset['embark_town']=le.fit_transform(dataset['embark_town'])
dataset['class']=le.fit_transform(dataset['class'])
dataset['alive']=le.fit_transform(dataset['alive'])
dataset['alone']=le.fit_transform(dataset['alone'])
dataset['who']=le.fit_transform(dataset['who'])
dataset.head(5)
```

```python
dataset.drop('deck', axis=1, inplace=True)
```

```python
corr = dataset.corr()
sns.heatmap(corr)
```

```python
sns.heatmap(corr,annot=True, cmap='crest_r')
```
